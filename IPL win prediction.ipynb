```python
import pandas as pd
import numpy as np
```

```python
delivery= pd.read_csv('deliveries.csv')
match=pd.read_csv('matches.csv')

match
```

```python
delivery.info()
```

```python
# in delivery dataset we extract only 3 col.
total_score=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()# covert into dataframe.reset_index()
```

```python
total_score
```

```python
total_score=total_score[total_score['inning']==1]
```

```python
# we merge the total_score into match df 
match_df=match.merge(total_score[['match_id','total_runs']],left_on='id',right_on='match_id')# id and match id basis we sort
```

```python
match_df
```

```python
# some teams are not playing like pune and gujrat
match_df['team1'].unique()
```

```python
#current playing teams
team=[    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

```

```python
match_df = match_df[match_df['team1'].isin(team)]
match_df = match_df[match_df['team2'].isin(team)]# print the team name which are present in team list
```

```python
match_df.shape
```

```python
match_df
```

```python
match_df['dl_applied'].value_counts() # rain affected
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df
```

```python
# here total_run_y says current score on each ball, mean 1st ball 1 run then 2nd ball o run
delivery_df['current_score']=delivery_df.groupby('match_id').cumsum()['total_runs_y']# cumsum mean add value with previous value
```

```python
delivery_df
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
#balls left
delivery_df['balls_left']= 126-(delivery_df['over']*6+delivery_df['ball']) # every match has 120 balls
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# current run rare(crr)= run/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# required run rate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
def result( row):
    return 1 if row['batting_team']==row['winner'] else 0 
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
# shuffling
final_df = final_df.sample(final_df.shape[0])
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
#final_df=final_df[final_df['balls_left']!=0]
final_df = final_df[final_df['balls_left'] != 0]
```

```python
#X=final_df.iloc[:,:-1]
#y=final_df.iloc[:,-1]
#from sklearn.model_selection import train_test_split
#X_train,y_train,X_test,y_test=train_test_split(X,y,test_size=0.2,random_state=32)

X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
# 1st 3 col. is string so, we do ohe
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]    
,remainder= 'passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
# value error:Input X contains infinity or a value too large for dtype('float64').
final_df.describe() #  in rrr max and min is infinite to -inf. so it occurs when we have runs but there is no ball ,eg 1/0
# so we delete balls left having 0
```

```python
y_pred=pipe.predict(X_test)
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
